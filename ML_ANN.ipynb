```python
dataset =  pd.read_csv("/content/drive/MyDrive/Second Sem/data/Bank_data.csv")
dataset.head()
```

```python
X = dataset.iloc[:,3:13].values
y = dataset.iloc[:,13].values
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelEncoder_X1 = LabelEncoder()
X[:,1] = labelEncoder_X1.fit_transform(X[:,1])
X
```

```python
labelEncoder_X2 = LabelEncoder()
X[:,2] = labelEncoder_X2.fit_transform(X[:,2])
X
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("Geography", OneHotEncoder(),[1])], remainder = "passthrough")
X = ct.fit_transform(X)
X.astype(int)
X
```

```python
X = X[:,1:]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
# ANN model is a sequense of layes - feedforward model
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initialising ANN
classifier = Sequential()
```

```python
#adding input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim =11))
```

```python
#adding the second hidden layer
#relu - rectified linear unit
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
#output layer - for multiple outputs use activation function softmax
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
#compiling the ANN - adam optimizes weights using Stochastic gradient descent
# loss function for binary classification is binary_crossentropy
#loss function for multi-class classification model where there are two or more output labels categorical_crossentropy
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#fitting the ANN to the traing set
classifier.fit(X_train, y_train, batch_size=10, epochs=10)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred>0.5)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
