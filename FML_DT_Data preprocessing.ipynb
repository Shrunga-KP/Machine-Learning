```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/Social_Network_Ads.csv")
dataset = data
dataset
```

```python
X = dataset.iloc[:,[2,3]].values
```

```python
y = dataset.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn import metrics
prediction = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", prediction * 100, '%')
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_test[:, 0].min() -1, stop = X_set[:,0].max() +1, step = 0.01),
                     np.arange(start = X_test[:, 1].min() -1, stop = X_set[:,1].max() +1, step = 0.01))
plt.contourf(X1,X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmp = ListedColormap(('black','brown')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i, j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set == j,0], X_set[y_set == j,1],
              c= ListedColormap(('red','green'))(i), label=j)
  plt.title('Decison tree classification (training set)')
  plt.xlabel('Age')
  plt.ylabel('Estimated salary')
  plt.legend()
  plt.show()

```
