```python
data1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Data/Wine.csv")
data1
```

```python
x = data1.iloc[: , 0:13].values
y = data1.iloc[: , 13].values
```

```python
#splitting data into the training and test

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#applying PCA

from sklearn.decomposition import PCA
#pca = PCA(n_components = None)
pca = PCA(n_components = 2)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
#fit Logistic Regression for the training set

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
#

y_pred = classifier.predict(x_test)
y_pred
```

```python
#making the confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#find accuracy

from sklearn import metrics
prediction = metrics.accuracy_score(y_test,y_pred)
print("Accuracy",prediction *100,'%')
```

```python
#predict the result of the training set

from matplotlib.colors import ListedColormap

x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[: , 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[: , 1].min() - 1, stop = x_set[: , 1].max() + 1, step = 0.01))

plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape), alpha = 0.75,
             cmap = LCM (('black', 'grey', 'white')))

plt.xlim (x1.min(), x1.max())
plt.ylim (x2.min(), x2.max())

for s, t in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == t, 0], x_set[y_set == t, 1],
                c = ListedColormap (('grey', 'black', 'blue'))(s), label = t)

plt.title('Logistic Regression for Training set ')
plt.xlabel ('PC1') # for X_label
plt.ylabel ('PC2') # for Y_label
plt.legend() # for showing legend

# show scatter plot
plt.show()
```

```python
#visualize the result of the testing set

from matplotlib.colors import ListedColormap as LCM

x_set, y_set = x_test, y_test

x1, x2 = np.meshgrid(np.arange(start = x_set[: , 0].min() - 1, stop = x_set[: , 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[: , 1].min() - 1, stop = x_set[: , 1].max() + 1, step = 0.01))

plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(),  x2.ravel()]).T).reshape(x1.shape), alpha = 0.75,
             cmap = LCM(('pink', 'grey', 'white')))

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for s, t in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == t, 0], x_set[y_set == t, 1],
                c = LCM(('red', 'black', 'blue'))(s), label = t)

# title for scatter plot
plt.title('Logistic Regression for Testing set')
plt.xlabel ('PC1') # for X_label
plt.ylabel ('PC2') # for Y_label
plt.legend()

# show scatter plot
plt.show()
```
