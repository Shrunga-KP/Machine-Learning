```python
import numpy as np
```

```python
#mp-neuron model for AND
def mcculloch_pitts_and(x1, x2):
    w1 = 1
    w2 = 1
    threshold = 2
    weighted_sum = x1 * w1 + x2 * w2
    output = 1 if weighted_sum >= threshold else 0
    return output

inputs =np.array( [[0, 0], [0, 1], [1, 0], [1, 1]])
for x1, x2 in inputs:
    result = mcculloch_pitts_and(x1, x2)
    print(f"Input: ({x1}, {x2}) -> Output: {result}")
```

```python
#mp-neuron model for OR
def mcculloch_pitts_or(x1, x2):
    w1 = 1
    w2 = 1
    threshold = 1
    weighted_sum = x1 * w1 + x2 * w2
    output = 1 if weighted_sum >= threshold else 0
    return output

inputs =np.array( [[0, 0], [0, 1], [1, 0], [1, 1]])
for x1, x2 in inputs:
    result = mcculloch_pitts_or(x1, x2)
    print(f"Input: ({x1}, {x2}) -> Output: {result}")
```

```python
#mp-neuron model for NOR
def mcculloch_pitts_nor(x1, x2):
    w1 = 1
    w2 = 1
    threshold = 1
    weighted_sum = x1 * w1 + x2 * w2
    output = 1 if weighted_sum < threshold else 0
    return output

inputs =np.array( [[0, 0], [0, 1], [1, 0], [1, 1]])
for x1, x2 in inputs:
    result = mcculloch_pitts_nor(x1, x2)
    print(f"Input: ({x1}, {x2}) -> Output: {result}")
```

```python
#mp-neuron model for NAND
def mcculloch_pitts_nand(x1, x2):
    w1 = 1
    w2 = 1
    threshold = 1
    weighted_sum = x1 * w1 + x2 * w2
    output = 1 if weighted_sum <= threshold else 0
    return output

inputs =np.array( [[0, 0], [0, 1], [1, 0], [1, 1]])
for x1, x2 in inputs:
    result = mcculloch_pitts_nand(x1, x2)
    print(f"Input: ({x1}, {x2}) -> Output: {result}")
```

```python
#mp-neuron model for NOT
def mcculloch_pitts_not(x):
    w=1
    threshold = 1
    weighted_sum = x*w
    output = 1 if weighted_sum < threshold else 0
    return output

inputs =[(0),(1)]
for x in inputs:
    result = mcculloch_pitts_not(x)
    print(f"Input: ({x}) -> Output: {result}")
```
